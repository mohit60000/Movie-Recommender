```python
#Loading the preprocessed complete movie data
import pandas as pd
complete_movies_preprocessed_data=pd.read_csv('complete_movies_processed_data.csv/part-00000',sep='::',names = ["movie_id", "title", "year", "genres"])
```

```python
#Removing the movies whose release year was not available
import numpy as np
data_columns = ['year']
complete_movies_preprocessed_data = (complete_movies_preprocessed_data.drop(data_columns, axis=1)
         .join(complete_movies_preprocessed_data[data_columns].apply(pd.to_numeric, errors='coerce')))
```

```python
#Converting the values of the column year to int
complete_movies_preprocessed_data=complete_movies_preprocessed_data[complete_movies_preprocessed_data[data_columns].notnull().all(axis=1)]
complete_movies_preprocessed_data['year']=complete_movies_preprocessed_data['year'].apply(lambda x:int(x))
```

```python
complete_movies_preprocessed_data
```

```python
#Listing movies with genres listed as '(no genres listed)'
print ("Movies with 'genres' listed as '(no genres listed)' are ",len(complete_movies_preprocessed_data[complete_movies_preprocessed_data['genres']=='(no genres listed)']))
```

```python
#removing the movies with genres listed as '(no genres listed)'
complete_movies_preprocessed_data=complete_movies_preprocessed_data[~(complete_movies_preprocessed_data['genres']=='(no genres listed)')]
```

```python
#Saving the cleaned data in csv file
complete_movies_preprocessed_data.to_csv('complete_movies_cleaned.csv',sep='%')
```

```python
#Loading the preprocessed complete ratings data
complete_ratings_preprocessed_data=pd.read_csv('complete_ratings_processed_data.csv/part-00000',sep='%',names = ['user_id','movie_id','rating','timestamp'])
```

```python
#Converting the type of values in movie_id to int and storing the number of ratings per movie id in ratings_movie_id_len
complete_ratings_preprocessed_data['movie_id']=complete_ratings_preprocessed_data['movie_id'].astype(int)
ratings_movie_id_len=complete_ratings_preprocessed_data.groupby('movie_id').size()
complete_ratings_preprocessed_data
```

```python
#Getting the user id who have rated just 1 movie
one_rating_index=[i+1 for i, x in enumerate(ratings_movie_id_len) if x == 1]
len(one_rating_index)
```

```python
#Getting movie ids which do have more than 1 ratings
movie_id_set = set(complete_ratings_preprocessed_data.movie_id)
one_rating_index_set = set(one_rating_index)
print(len(movie_id_set),len(one_rating_index_set),len(movie_id_set-one_rating_index_set))
result_set=movie_id_set-one_rating_index_set
```

```python
#Storing just the movies with more than 1 ratings and removing the rest
complete_ratings_cleaned=complete_ratings_preprocessed_data[complete_ratings_preprocessed_data.movie_id.isin(list(result_set))]
```

```python
#Saving the cleaned data
complete_ratings_cleaned.to_csv('complete_ratings_cleaned.csv',sep='%')
```
